# Problem 1: Social Media – Post Engagement Analyzer


```python
def analyze_engagement(likes):
    total_likes = 0

    for like in likes:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


likes_list = [200, 150, 300, 250, 150]

analyze_engagement(likes_list)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def check_rainfall(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


season_rainfall = [60, 80, 75, 70, 75]
required = 70

check_rainfall(season_rainfall, required)
```

# Problem 4: Social Media – Duplicate Account Detection


```python
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(usernames) != len(unique_users):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


user_list = ["user1", "user2", "user3", "user1", "user4"]

detect_duplicates(user_list)
```

# Problem 6: Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


crop_prices = [1500, 2500, 1800, 3200, 1900]

filter_premium_crops(crop_prices)
```

# Problem 7: System Monitoring – Application Health Checker


```python
def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)


error_count = int(input("Enter Error Count: "))

check_system_health(error_count)
```

# Problem 8: Banking – Daily Transaction Limit Checker


```python
def check_transaction_limit(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


transaction_amount = int(input("Enter Transaction Amount: ₹"))

check_transaction_limit(transaction_amount)
```

# Problem 9: E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


attendance_list = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]

check_exam_eligibility(attendance_list)
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


units_consumed = int(input("Enter Units Consumed: "))

total_bill, usage_status = calculate_bill(units_consumed)

print("Total Electricity Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
