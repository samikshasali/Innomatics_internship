Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")

sentence = sentence.lower()

words = sentence.split()

unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data


```python
# Given dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Display result
print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

Problem 3: Find Maximum and Minimum Values


```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
max_value = numbers[0]
min_value = numbers[0]

# Compare each element
for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

# Print results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Problem 4: Count Products Above a Price Threshold


```python
# Given list of prices
prices = [450, 1200, 899, 1500, 300]

count = 0

# Check each price
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage


```python
# Given attendance list
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

# Count "P"
for day in attendance:
    if day == "P":
        present_count += 1

# Calculate percentage
total_days = len(attendance)
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

Problem 7: Count Character Frequency


```python
# Given string
text = "pythonp"

# Create empty dictionary
char_count = {}

# Count character frequency
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

# Print result
print(char_count)
```

Problem 9: Check if a Key Exists in a Dictionary


```python
# Given dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Problem 10: Calculate Average Marks


```python
# Given list of marks
marks = [65, 75, 80, 70]

# Calculate average
total = sum(marks)
count = len(marks)
average = total / count

# Print result
print("Average Marks:", average)
```
