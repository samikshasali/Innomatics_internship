# Problem 1: Mobile Recharge Validation System


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid Recharge Plan")
        return False
    else:
        return True


while True:
    recharge_amount = int(input("Enter Recharge Amount: RS."))

    if validate_recharge(recharge_amount):
        print("Recharge Successful")
        break
    else:
        print("Please try again!\n")
```

# Problem 2: Inventory Reorder Alert System


```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "→ Reorder Alert ")
        else:
            print(product, "→ Stock OK ")


products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(products)
```

# Problem 3: Student Result Processing System


```python
def process_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [60, 45, 70, 55, 40]

result = process_result(student_marks)

print("Average Marks:", sum(student_marks)/len(student_marks))
print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak == 'y':
        fare = fare + (0.25 * fare)

    return fare


while True:
    km = float(input("Enter Distance in KM: "))
    peak_hour = input("Is it Peak Hour? (y/n): ")

    total_fare = calculate_fare(km, peak_hour)

    print("Total Cab Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (y/n): ")
    if retry != 'y':
        print("Thank You")
        break
```

# Problem 5: Employee Attendance Eligibility Checker


```python
def check_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

result = check_eligibility(attendance_list)

print("Attendance Percentage:", (attendance_list.count("P")/len(attendance_list))*100)
print("Eligibility Status:", result)
```

# Problem 6: Password Strength Checker


```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password "

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password "


pwd = input("Enter Password: ")

result = check_password(pwd)

print(result)
```
