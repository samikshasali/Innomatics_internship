# Problem 1: Smart Light Controller


```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"

    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"

    # Method to display status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Taking user input
name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

# Creating object
light1 = SmartLight(name)

# Performing action
if action.upper() == "ON":
    light1.turn_on()
elif action.upper() == "OFF":
    light1.turn_off()
else:
    print("Invalid Action!")

# Displaying status
light1.display_status()
```

# Problem 2: Employee ID Card System


```python
class Employee:

    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card details
    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Taking user input
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
dept = input("Enter Department: ")

# Creating object
emp1 = Employee(name, emp_id, dept)

# Displaying ID Card
emp1.display_id_card()
```

# Problem 3: Mobile Contact Record


```python
class Contact:

    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact details
    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Taking user input
name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

# Creating object
c1 = Contact(name, phone)

# Displaying contact info
c1.display_contact()
```

# Problem 4: Product Price Tag Generator


```python
class Product:

    # Constructor
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    # Method to print formatted price tag
    def display_price_tag(self):
        print("\nProduct:", self.product_name)
        print("Price: ₹" + str(self.price))


# Taking user input
name = input("Enter Product Name: ")
price = int(input("Enter Price: "))

# Creating object
p1 = Product(name, price)

# Displaying price tag
p1.display_price_tag()
```

# Problem 5: Movie Rating Display System


```python
class Movie:

    # Constructor
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print("\nMovie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


# Taking user input
name = input("Enter Movie Name: ")
rating = float(input("Enter Rating: "))

# Creating object
m1 = Movie(name, rating)

# Displaying movie details
m1.display_movie()
```

# Problem 6: Delivery Address Manager


```python
class Delivery:

    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_delivery(self):
        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Taking user input
name = input("Enter Customer Name: ")
addr = input("Enter Address: ")

# Creating object
d1 = Delivery(name, addr)

# Displaying delivery details
d1.display_delivery()
```
