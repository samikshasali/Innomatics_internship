# Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):

    current_vehicles = 0
    peak_vehicles = 0

    for log in logs:

        if log == "IN":
            current_vehicles += 1

        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_vehicles:
            peak_vehicles = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_vehicles)

    if current_vehicles > capacity:
        print(" ALERT: Parking Exceeded Capacity!")
    else:
        print("Parking Status: Available")
```

```python
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

# Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic, weather):

    # Step 1: Base delivery time
    base_time = (distance / 20) * 60

    delay = 0

    # Step 2: Traffic delay
    if traffic == "Medium":
        delay += 10
    elif traffic == "High":
        delay += 20

    # Step 3: Weather delay
    if weather == "Rainy":
        delay += 15
    elif weather == "Stormy":
        delay += 25

    # Step 4: Final ETA
    eta = base_time + delay

    print("Estimated Delivery Time:", int(eta), "minutes")
```

```python
estimate_delivery_time(8, "High", "Rainy")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy(total_seats, booked_seats):

    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")

    elif occupancy >= 75:
        print("Show Status: Almost Full")

    else:
        print("Show Status: Available")

    if occupancy >= 90:
        print("Suggestion: Open Additional Show")
```

```python
total = 200
booked = [1] * 150

seat_occupancy(total, booked)
```

# Problem Statement 4:Cloud Server Load Classification System


```python
def server_load(cpu_readings):

    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")

    elif average <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")
```

```python
cpu = [45, 60, 70, 85, 90]

server_load(cpu)
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def resource_monitor(usage):

    overused = []

    for resource in usage:

        if usage[resource] > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ",".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
```

```python
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(resources)
```

# Problem Statement 6: Online Event Registration Capacity Controller


```python
def event_registration(capacity, registrations):

    confirmed = 0
    waitlist = 0

    for i in range(registrations):

        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    if waitlist > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")
```

```python
event_registration(100, 105)
```
