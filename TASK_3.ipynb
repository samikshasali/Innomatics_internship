# Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())

top_performers = []
for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

# Problem Statement 2: Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online"

query = query.lower()

import string
query = query.translate(str.maketrans("", "", string.punctuation))

words = query.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

result = {}

for word in word_count:
    if word_count[word] > 1:
        result[word] = word_count[word]

print(result)
```

# Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index in range(len(sensor_readings)):
    if sensor_readings[index] % 2 == 0:
        valid_readings.append((index, sensor_readings[index]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_users = len(emails)

print("Domain Usage Analysis:\n")

for domain in domain_count:
    percentage = (domain_count[domain] / total_users) * 100
    print(f"{domain} → {domain_count[domain]} users ({percentage:}%)")
```

# Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)

threshold = average_sales * 1.30

for index in range(len(sales)):
    if sales[index] > threshold:
        print(f"Day {index + 1}: {sales[index]}")
```

# Problem Statement 6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

for user in id_count:
    if id_count[user] > 1:
        print(f"{user} → {id_count[user]} times")
```
